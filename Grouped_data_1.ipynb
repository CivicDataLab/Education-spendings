```python
import numpy as np
import pandas as pd # calling the important libraries
```

```python
spen1=pd.read_csv("district_spending_expenditure_fy_2018_19.csv") # reading the dataset
```

```python
spen2=pd.read_csv("district_spending_expenditure_fy_2019_20_lat.csv")
```

```python
spen1.head() # displaying first 5 rows of spending data of fiscal year 2018_19
```

```python
spen2.head() # displaying first 5 records of fiscal year 2019_20
```

```python
spen1=spen1.groupby(['demand','demand_desc', 'major', 'major_desc', 'sub_major',
       'sub_major_desc', 'minor', 'minor_desc', 'sub_minor', 'sub_minor_desc', 'SOE', 'SOE_description'])["NETPAYMENT"].aggregate("mean")
spen1.head() # grouping the dataset by certain columns and finding mean of netpayment 
```

```python
spen2=spen2.groupby(['demand','demand_desc', 'major', 'major_desc', 'sub_major',
       'sub_major_desc', 'minor', 'minor_desc', 'sub_minor', 'sub_minor_desc', 'SOE', 'SOE_description'])["NETPAYMENT"].aggregate("mean")
spen2.head()
```

```python
spen1.to_excel("spendings_fy_2018_19.xlsx",merge_cells=False) # converting into excel files
```

```python
spen2.to_excel("spendings_fy_2019_20.xlsx",merge_cells=False)
```
